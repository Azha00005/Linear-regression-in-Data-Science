# *Linear Regression:* y=kx+b (k=tgA)

$$h(x) = w_0+w_1*x$$

```python
def h(w_0, w_1, x):
  return w_0+w_1*x
```

# *LOSS function*

```python
def loss_funtion(w_0, w_1, df):
  n=df.area.shape[0]
  cost=0

  for x, y in zip(df.area, df.price):
    cost = cost + (h(w_0, w_1, x)-y)**2

  return cost/(2*n)
```

# ***Gradient Descent***

```python
def grad_step(weights, grads, learning_rate=0.001):
  w_0, w_1 = weights
  grad_0, grad_1 = grads

  w_0 = w_0 - learning_rate*grad_0
  w_1 = w_1 - learning_rate*grad_1

  return w_0, w_1
```

```python
def grad_w_0(w_0, w_1, df):
  n = df.area.shape[0]
  grad = 0

  for x, y in zip(df.area, df.price):
    grad = grad + (h(w_0, w_1, x)-y)

  return grad/n

def grad_w_1(w_0, w_1, df):
  n = df.area.shape[0]
  grad = 0

  for x, y in zip(df.area, df.price):
    grad = grad + (h(w_0, w_1, x)-y)*x

  return grad/n
```

```python
def grad_descent(weights, df, num_iter, learning_rate=0.001, epsilon = 0.01):
  w_0, w_1 = weights
  loss=loss_funtion(w_0, w_1, df)
  loss_history = [loss]

  for i in range(num_iter):
    d_w_0 = grad_w_0(w_0, w_1, df)
    d_w_1 = grad_w_1(w_0, w_1, df)
    grads = (d_w_0, d_w_1)
    #gradient computetion
    w_0, w_1 = grad_step(weights, grads, learning_rate= learning_rate)
    weights = (w_0, w_1)
    loss = loss_funtion(w_0, w_1, df)

    if abs(loss-loss_history[-1]) < epsilon:
      loss_history.append(loss)
      break

    loss_history.append(loss)

  return weights, loss_history
```

**Normalization**

```python
def normalization(data):
  mean = np.mean(data)
  value_range = np.max(data) - np.min(data)
  result =[]

  for x in data:
    norm_x = (x-mean)/value_range
    result.append(norm_x)

  return result
```
